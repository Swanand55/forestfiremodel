# Simple Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('height_weight.csv')
```

```python
df.head()
```

```python
plt.scatter(df['weight'],df['height'])
plt.xlabel('weight')
plt.ylabel('height')
plt.show()
```

```python
## divide our dataset into dep and indep features
```

```python
X = df[['weight']] ## independent feature
y = df[['height']] ## dependent feature
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
# standardize independent data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train) #z-score application
X_test_scaled = scaler.transform(X_test)
```

```python
plt.scatter(X_train, y_train)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print('Slope : ',linreg.coef_, ' and  Intercept : ', linreg.intercept_)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, linreg.predict(X_train), 'r')
plt.show()
```

```python
y_pred_test = linreg.predict(X_test)
y_test , y_pred_test
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, linreg.predict(X_test), 'r')
plt.show()
```

# performance metrics
## mse, mae, rmse
## r2, adjusted r2

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred_test)
mae = mean_absolute_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
```

```python
print("MAE : ", mae, ", MSE : ", mse, ", RMSE : ", rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred_test)
score
```

## Adjusted R Square

```python
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
